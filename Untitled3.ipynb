```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
class ConcreteMLP(nn.Module):
    def __init__(self,input_size=8,hidden_size=16,output_size=1):
        super(ConcreteMLP,self).__init__()

        #define the layer and the number of neuron in each layers
        self.fc1=nn.Linear(input_size,hidden_size)
        self.relu=nn.ReLU()
        self.fc2=nn.Linear(hidden_size,output_size)

    def forward(self,x):
        x=self.fc1(x)
        x=self.relu(x)
        x=self.fc2(x)
        return x

```

```python
data=pd.read_excel("/content/Concrete_Data.xls")
inputs=data.iloc[:,:-1].values
output=data.iloc[:,-1].values.reshape(-1,1)

#standardize the input data
scaler=StandardScaler()
input_transformed=scaler.fit_transform(inputs)

#convert into tensor
input_tensor=torch.tensor(input_transformed, dtype=torch.float32)
output_tensor=torch.tensor(output, dtype=torch.float32)

#split the data
torch.manual_seed(42)
n_of_feat=input_tensor.shape[0]
indices=torch.randperm(n_of_feat)

#define the cutoff
cutoff=int(0.8*n_of_feat)
train_indices=indices[:cutoff]
test_indices=indices[cutoff:]

X_train=input_tensor[train_indices]
y_train=output_tensor[train_indices]

X_test=input_tensor[test_indices]
y_test=output_tensor[test_indices]

print(X_train.shape)
print(y_train.shape)
```

```python

plt.hist(output_tensor,edgecolor="k")
plt.xlabel("Concrete strength")
plt.ylabel("Count")
plt.show()
```

```python

plt.scatter(data['Cement (component 1)(kg in a m^3 mixture)'],output,alpha=0.5)
plt.xlabel("Concrete strength")
plt.ylabel("Cement")
plt.show()
```

```python
model=ConcreteMLP()
mse_loss=nn.MSELoss()
optimizer=optim.Adam(model.parameters(),lr=0.01)
print(model)
print(optimizer)
```

```python
epochs=300
train_losses=[]
for epoch in range(epochs):

    y_pred=model(X_train)
    loss=mse_loss(y_pred,y_train)
    loss.backward()

    optimizer.step()

    optimizer.zero_grad()
    train_losses.append(loss.item())

    if(epoch+1)%50==0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
fig,ax=plt.subplots(figsize=(8,5))
plt.plot(range(epochs),train_losses,linestyle='-', color='blue')
plt.xlabel('Epoch')
plt.ylabel('Training Loss (MSE)')
plt.title('Training Loss Over Time')
plt.grid(True)
plt.show()
```

```python
with torch.no_grad():
    train_pred=model(X_train)
    train_loss=mse_loss(train_pred,y_train)

    test_pred=model(X_test)
    test_loss=mse_loss(test_pred,y_test)

print(f"Final Training Loss (MSE): {train_loss.item():.4f}")
print(f"Final Testing Loss (MSE): {test_loss.item():.4f}")
```

```python
# Calculate RMSE
train_rmse = torch.sqrt(train_loss)
test_rmse = torch.sqrt(test_loss)

print(f"Training RMSE: {train_rmse.item():.2f} MPa")
print(f"Testing RMSE: {test_rmse.item():.2f} MPa")
```
